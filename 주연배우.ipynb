```python
from bs4 import BeautifulSoup 
from urllib.request import urlopen

import pandas as pd
import re
```

```python
df=pd.read_csv('./df.csv', encoding='EUCKR')
df=df.drop(['Unnamed: 0'], axis=1)
```

```python
df
```

```python
#주연배우
Actor=[]
for i in range(len(df['제목'])):
    url_page = "https://movie.naver.com/movie/bi/mi/basic.nhn?code={}".format(df['코드'][i])
    html = urlopen(url_page)
    soup = BeautifulSoup(html, "lxml")
    souplist=soup.find_all('dd')
    souplist[1].get_text()
    actor=souplist[1].get_text().replace('\n','').replace('\t','').replace('\r','').replace(' ','').split(',')
    actor2=[]
    for item in actor:
        temp1=re.sub('\([가-힝A-Za-z]+\)','',item)
        temp2=re.sub('\([가-힝A-Za-z]+\)','',temp1)
        actor2.append(temp2)
    Actor.append(actor2)
```
